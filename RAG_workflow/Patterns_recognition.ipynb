```python
def process_strings(input_list):
    """
    Takes a list of strings, separates strings containing a semicolon into two,
    and returns a set of all unique strings obtained.

    Args:
        input_list (list): A list of strings.

    Returns:
        set: A set of unique strings.
    """
    result = set()
    for string in input_list:
        # Check if the string contains a semicolon
        if ';' in string:
            # Split by semicolon and add both parts to the result set
            parts = string.split(';')
            parts_stripped = [x.strip() for x in parts]
            result.update(parts_stripped)
        else:
            # Add the string as is
            result.add(string.strip())
    return result

# Example usage

result_set = process_strings(input_list)
print("Result set:", result_set)
```
